# Mooney Reconstruction
I will try to restore photos of celebrities from Mooney photos, which are binarized faces.
In order to do this, I will leverage a large training set of grayscale faces and Mooney faces.

Producing a face reconstruction from a binarized counterpart is a challenging high dimensional problem, but I will show that I can learn to do so from data. In particular, using the power of Canonical Correlation Analysis (CCA), I will reduce the dimensionality of the problem by projecting the data into a subspace where the images are most correlated

# Part 1: Covariance Matrix Computation
The following datasets will be used for this project: 
* X train.p: contains 956 binarized images, where each image Xi ∈ R15×15×3.
* Y train.p: contains 956 corresponding grayscale images, where Yi ∈ R15×15×3.
* X test.p: contains 255 binarized images, where each image Xi ∈ R15×15×3.
* Y test.p: contains 255 corresponding grayscale images,where Yi ∈ R15×15×3.

```python
import os
import numpy as np
import cv2
import copy
import glob
import sys
from numpy.random import uniform
import pickle
from scipy.linalg import eig
from scipy.linalg import sqrtm
from numpy.linalg import inv
from numpy.linalg import svd
import numpy.linalg as LA
import matplotlib.pyplot as plt
import IPython
from sklearn.preprocessing import StandardScaler
import pdb
```


Through out the problem we will assume that all data points are flattened and standardize as follows:

x -> (x/255)∗2.0−1.0 and y -> (y/255)∗2.0−1.0.

In order to use CCA to find pairs of directions a and b that maximize the correlation between X train and Y train, fitst I need to estimate the three covariance matrices from finite samples of data.

In order to properly compute the covariance matrices you have to 1) standardize the data, 2) subtract the mean and 3) scale by the number of data points at the end.

```python
def standardized(v):
    return (v/255.0) * 2.0 - 1.0

def flatten_and_standardize(data):
    result = []
    for d in data:
      d = d.flatten()
      d = standardized(d)
      result.append(d)
    return result

class Mooney(object):

    def __init__(self):
        
        #For numerical issues we need to add a very small scalar times the identity matrix to the covariance terms.
        self.lmbda = 1e-5

    def load_data(self):
        self.x_train = pickle.load(open('x_train.p','rb'))
        self.y_train = pickle.load(open('y_train.p','rb'))
        self.x_test = pickle.load(open('x_test.p','rb'))
        self.y_test = pickle.load(open('y_test.p','rb'))

    def compute_covariance_matrices(self):
        # use StandardScaler to do mean subtraction
        ss_x = StandardScaler(with_std = False)
        ss_y = StandardScaler(with_std = False)

        num_data = len(self.x_train)

        # read the first trainning data
        x = self.x_train[0]
        y = self.y_train[0]

        # flatten the data
        x_f = x.flatten()
        y_f = y.flatten()

        # keep record of the dimention of the data
        x_f_dim = x_f.shape[0]
        y_f_dim = y_f.shape[0]
        self.x_dim = x_f_dim
        self.y_dim = y_f_dim

        # Calculate the covariance matrix of xx, yy and xy
        self.C_xx = np.zeros([x_f_dim,x_f_dim])
        self.C_yy = np.zeros([y_f_dim,y_f_dim])
        self.C_xy = np.zeros([x_f_dim,y_f_dim])

        x_data = []
        y_data = []

        # loop through data to calculate the sum of (xi-xmean)(xi-xmean).T
        for i in range(num_data):
            x_image = self.x_train[i]
            y_image = self.y_train[i]

            # FLATTEN DATA
            x_f = x_image.flatten()
            y_f = y_image.flatten()

            # STANDARDIZE DATA
            x_f = standardized(x_f)
            y_f = standardized(y_f)

            x_data.append(x_f)
            y_data.append(y_f)

        # SUBTRACT MEAN
        ss_x.fit(x_data)
        x_data = ss_x.transform(x_data)

        ss_y.fit(y_data)
        y_data = ss_y.transform(y_data)

        for i in range(num_data):
            x_f = np.array([x_data[i]])
            y_f = np.array([y_data[i]])
            # COMPUTE COVARIANCE MATRICES
            self.C_xx += x_f.T @ x_f
            self.C_yy += y_f.T @ y_f
            self.C_xy += x_f.T @ y_f

        # DIVIDE BY THE NUMBER OF DATA POINTS
        self.C_xx = 1.0/float(num_data)*self.C_xx
        self.C_yy = 1.0/float(num_data)*self.C_yy
        self.C_xy = 1.0/float(num_data)*self.C_xy
```

```python
mooney = Mooney()
mooney.load_data()
mooney.compute_covariance_matrices()
```

# Part 2: Explore Singular Value Decay
CCA looks for the maximum singular value of the matrix ΣXX^{−1/2}ΣXY ΣYY^{−1/2}, and that this corresponds to the maximum correlation coefficient. For numerical issues we need to add a very small scalar lambda times the identity matrix to the covariance terms. 

```python
class Mooney(Mooney):
          
    def solve_for_variance(self, plot=False):
        # calculate the sigular values
        
        eigen_values = np.zeros((675,))
        eigen_vectors = np.zeros((675, 675))

        # compute C_xx_inv_sqrt  and C_yy_inv_sqrt
        self.C_xx_inv_sqrt = inv(sqrtm(self.C_xx + 0.00001 * np.identity(len(self.C_xx))))
        self.C_yy_inv_sqrt = inv(sqrtm(self.C_yy + 0.00001 * np.identity(len(self.C_yy))))

        # plot sigular value of the matrix
        temp = self.C_xx_inv_sqrt @ self.C_xy @ self.C_yy_inv_sqrt
        u, s, v = svd(temp)
        if plot:
            plt.clf()
            plt.plot(range(len(s)), s)
            plt.xlabel('features')
            plt.ylabel('singular values')
            plt.show()

        # compute correlation matrix
        evals, evecs = eig(temp @ temp.T)
        return evals, evecs
```

```python
mooney = Mooney()
mooney.load_data()
mooney.compute_covariance_matrices()
eig_val, eig_vec = mooney.solve_for_variance()
```

This plot clearly shows the singular value decay.It therefore makes sense to only consider the top singular value(s), as in CCA. 

# Part 3: Optimal Reduced Dimensions 
I use CCA as a technique to help learn better features for the problem.
I will use ridge regression regression to learn a mapping, W ∈ k×675 (15 ∗ 15 ∗ 3 = 675), from the projected binarized data to the grayscale images. 

Here the reduced dimension k is the key parameter for our model.
Too small k will miss some important features, while too large k will overfit the model.
To find the optimal k, I plot the test error for different k and choose k with the minimum test error.

```python
class Mooney(Mooney):
    def project_data(self, eig_val,eig_vec,proj=150):
        # compute projection siguar vectors
        return eig_vec.T[0:proj].T

    def ridge_regression(self):
        # implement ridge regression
        w_ridge = inv(self.X_ridge.T @ self.X_ridge + 0.00001 * np.identity(len(self.X_ridge[0]))) @ self.X_ridge.T @ self.Y_ridge
        self.w_ridge = w_ridge

    def compute_projected_data_matrix(self,X_proj):
        # compute the matrix of projected data
        Y = []
        X = []
        Y_test = []
        X_test = []
     
        num_data = len(self.x_train)

        # load trainning data
        for x in self.x_train:
            x_f = np.array([x.flatten()])
            # standarized data 
            x_f = standardized(x_f)
            # project data 
            X.append((x_f @ X_proj)[0])
        Y = flatten_and_standardize(self.y_train)

        # load test data
        for x in self.x_test:
            x_f = np.array([x.flatten()])
            x_f = standardized(x_f)
            X_test.append((x_f @ X_proj)[0])
        Y_test = flatten_and_standardize(self.y_test)

        # convert to matrix
        self.X_ridge = np.vstack(X)
        self.Y_ridge = np.vstack(Y)

        self.X_test_ridge = np.vstack(X_test)
        self.X_test_ridge = X_test
        self.Y_test_ridge = np.vstack(Y_test)

        return X

    def measure_error(self, X_ridge, Y_ridge):
        # calculate the error using projected data
        prediction = X_ridge @ self.w_ridge
        evaluation = Y_ridge - prediction

        dim,num_data = evaluation.shape

        error = []

        for i in range(num_data):
          # compute L2 norm for each vector then sqaure
          error.append(LA.norm(evaluation[:,i])**2)

        # Return average error
        return np.mean(error)
```

```python
mooney = Mooney()
mooney.load_data()
mooney.compute_covariance_matrices()
eig_val, eig_vec = mooney.solve_for_variance(plot=False)
proj = [1,50,100,150,200,250,300,350,400,450,500,650]
error_test = []
for p in proj:
    X_proj = mooney.project_data(eig_val,eig_vec,proj=p)

    # compute the regression
    mooney.compute_projected_data_matrix(X_proj)
    mooney.ridge_regression()
    training_error = mooney.measure_error(mooney.X_ridge, mooney.Y_ridge)
    test_error = mooney.measure_error(mooney.X_test_ridge, mooney.Y_test_ridge)
    error_test.append(test_error)

plt.clf()
plt.plot(proj,error_test)
plt.xlabel('k')
plt.ylabel('test error')
plt.show()
```

Squared Euclidean test error is minimized at k = 100, so I decide to reduce the trainning and test data to 100 features with largest eval.

# Part 4: Performing  Regression on Projected Data
I run the following ridge regression regression from the projected binarized data with k=10 to the grayscale images. Then I apply the learned model on 4 of the images in the test set and report the results.

```python
class Mooney(Mooney):
    def draw_images(self):
        plt.figure(figsize=(50,30))

        for count, x in enumerate(self.x_test):
            if ( count < 5 ):
                x = x.astype("uint8")
                x = cv2.resize(x,(100,100))
                plt.subplot(5,3,count*3+1)
                plt.imshow(x)
                if count == 4:
                    plt.xlabel('Mooney', fontsize=20)

        for count, x in enumerate(self.X_test_ridge):
            if ( count < 5): 
                prediction = np.matmul(self.w_ridge.T,x)
                prediction = ((prediction+1.0)/2.0)*255.0
                prediction = np.reshape(prediction,(15,15,3))
                p = prediction.astype("uint8")
                p = cv2.resize(p,(100,100))
                plt.subplot(5,3,count*3+2)
                plt.imshow(p)
                if count == 4:
                    plt.xlabel('Predicted', fontsize=20)
                    
        for count, x in enumerate(self.y_test):
            if ( count < 5 ):
                x = x.astype("uint8")
                x = cv2.resize(x,(100,100))
                plt.subplot(5,3,count*3+3)
                plt.imshow(x)
                if count == 4:
                    plt.xlabel('Ground Truth', fontsize=20)
        plt.tight_layout()
```
